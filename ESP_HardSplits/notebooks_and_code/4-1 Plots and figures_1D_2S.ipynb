```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import re

def parse_log(file_path):
    with open(file_path, 'r') as file:
        log_lines = file.readlines()

    iteration_data = []
    unique_losses = set()

    for line in log_lines:
        match_iteration = re.search(r'Iteration (\d+)', line)
        match_loss = re.search(r'Best loss so far: (\d+\.\d+)', line)

        if match_iteration:
            current_iteration = int(match_iteration.group(1))
        elif match_loss:
            current_loss = float(match_loss.group(1))

            # Check for unique loss in the same iteration
            if current_loss not in unique_losses:
                iteration_data.append({'iteration': current_iteration, 'loss': current_loss})
                unique_losses.add(current_loss)

    return pd.DataFrame(iteration_data)

log_directory = join(CURRENT_DIR, ".." ,"data","Reports","hyperOp_report")
log_files = [f for f in os.listdir(log_directory) if f.endswith('.log')]

color_map = {'ESP(1D)': 'red', 'C1e': 'blue', 'C1f': 'green', 'I1e': 'black', 'I1f': 'magenta'}

fig, ax = plt.subplots(figsize=(10, 6))

for log_file in log_files:
    experiment_name = os.path.splitext(log_file)[0].split('HOP_ESM1bts_and_ECFP_2S')[1]
    color = color_map.get(experiment_name, 'gray')
    log_data = parse_log(os.path.join(log_directory, log_file))
    
    # Smooth the data using moving average
    window_size = min(20, max(1, len(log_data['iteration']) // 10))  # Adjust window size as needed
    smooth_loss = np.convolve(log_data['loss'], np.ones(window_size)/window_size, mode='same')
    smooth_iteration = log_data['iteration']

    # Convert smooth_iteration and smooth_loss to Pandas Series
    smooth_iteration = pd.Series(smooth_iteration)
    smooth_loss = pd.Series(smooth_loss)

    # Extend the curve to iteration 2000
    last_iteration = smooth_iteration.iloc[-1]
    remaining_iterations = np.arange(last_iteration + 1, 2001)
    extension_loss = np.full_like(remaining_iterations, smooth_loss.iloc[-1])
    smooth_iteration = smooth_iteration.append(pd.Series(remaining_iterations))
    smooth_loss = smooth_loss.append(pd.Series(extension_loss))

    # Plot smoothed data
    ax.plot(smooth_iteration, smooth_loss, color=color, label=experiment_name, alpha=0.8)

    # Find minimum loss point from original data
    min_loss_index = log_data['loss'].idxmin()
    min_loss_iteration = log_data['iteration'].iloc[min_loss_index]
    min_loss = log_data['loss'].iloc[min_loss_index]
    
    # Mark minimum loss point
    ax.scatter(min_loss_iteration, min_loss, color=color, s=50, marker='o')
    
    # Annotate minimum loss value
    ax.annotate(f'{min_loss:.2f}', (min_loss_iteration, min_loss), textcoords="offset points", xytext=(5,5), ha='center', fontsize=11)

ax.set_xlabel('Iteration', fontsize=12)
ax.set_ylabel('Loss', fontsize=12)
ax.set_title('Loss per Iteration for ESM1bts + ECFP', fontsize=18, fontname='Arial')
ax.set_xlim(0, 2000)
#ax.grid(True)  # Add grid lines
ax.legend(loc='upper right', fontsize=10)  # Place legend outside the plot

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import re

def parse_log(file_path):
    with open(file_path, 'r') as file:
        log_lines = file.readlines()

    iteration_data = []
    unique_losses = set()

    for line in log_lines:
        match_iteration = re.search(r'Iteration (\d+)', line)
        match_loss = re.search(r'Best loss so far: (\d+\.\d+)', line)

        if match_iteration:
            current_iteration = int(match_iteration.group(1))
        elif match_loss:
            current_loss = float(match_loss.group(1))

            # Check for unique loss in the same iteration
            if current_loss not in unique_losses:
                iteration_data.append({'iteration': current_iteration, 'loss': current_loss})
                unique_losses.add(current_loss)

    return pd.DataFrame(iteration_data)

log_directory = os.path.join("..", "data", "Reports", "hyperOp_report")
log_files = [f for f in os.listdir(log_directory) if f.endswith('.log')]

color_map = {
    'ESP(1D)': 'red', 'C1e': 'blue', 'C1f': 'green', 'I1e': 'black', 'I1f': 'magenta',
    'PreGNN_C1e': 'blue', 'PreGNN_C1f': 'green', 'PreGNN_I1e': 'black', 'PreGNN_I1f': 'magenta',
    'ECFP_C1e': 'blue', 'ECFP_C1f': 'green', 'ECFP_I1e': 'black', 'ECFP_I1f': 'magenta',
    'PreGNN_ESPC1f': 'orange', 'ECFP_ESPC1f': 'orange', 'ECFP_C2': 'purple', 'PreGNN_C2': 'purple',
    'PreGNN_ESPC2': 'cyan', 'ECFP_ESPC2': 'cyan', 'PreGNN_ESP_NoATP': 'brown', 'ECFP_ESP_NoATP': 'brown',
    'PreGNN_ESP_D3408': 'pink', 'ECFP_ESP_D3408': 'pink', 'ECFP_ESPC1F': 'yellow'
}

# Separate log files into 2S and 3S, filtering to only those containing '1' in the experiment name
log_files_2S = [f for f in log_files if f.endswith('ECFP_2S.log') and re.search(r'1', f)]
log_files_3S = [f for f in log_files if f.endswith('ECFP_3S.log') and re.search(r'1', f)]

def plot_logs(log_files, title):
    fig, ax = plt.subplots(figsize=(10, 6))

    for log_file in log_files:
        experiment_name_match = re.search(r'HOP_ESM1bts_and_ECFP_(\w+)', log_file)
        if not experiment_name_match:
            continue  # Skip files that don't match the expected pattern

        experiment_name = experiment_name_match.group(2)
        if experiment_name not in color_map:
            print(f"Experiment name '{experiment_name}' not found in color_map. Using gray color.")
        color = color_map.get(experiment_name, 'gray')
        log_data = parse_log(os.path.join(log_directory, log_file))

        if log_data.empty:
            print(f"Log file {log_file} is empty or has no valid data.")
            continue

        if 'iteration' not in log_data.columns or 'loss' not in log_data.columns:
            print(f"Log file {log_file} does not contain expected columns.")
            continue

        # Smooth the data using moving average
        window_size = min(20, max(1, len(log_data['iteration']) // 10))  # Adjust window size as needed
        smooth_loss = np.convolve(log_data['loss'], np.ones(window_size) / window_size, mode='same')
        smooth_iteration = log_data['iteration']

        # Convert smooth_iteration and smooth_loss to Pandas Series
        smooth_iteration = pd.Series(smooth_iteration)
        smooth_loss = pd.Series(smooth_loss)

        # Extend the curve to iteration 2000
        last_iteration = smooth_iteration.iloc[-1]
        remaining_iterations = np.arange(last_iteration + 1, 2001)
        extension_loss = np.full_like(remaining_iterations, smooth_loss.iloc[-1])
        smooth_iteration = pd.concat([smooth_iteration, pd.Series(remaining_iterations)], ignore_index=True)
        smooth_loss = pd.concat([smooth_loss, pd.Series(extension_loss)], ignore_index=True)

        # Plot smoothed data
        ax.plot(smooth_iteration, smooth_loss, color=color, label=experiment_name, alpha=0.8)

        # Find minimum loss point from original data
        min_loss_index = log_data['loss'].idxmin()
        min_loss_iteration = log_data['iteration'].iloc[min_loss_index]
        min_loss = log_data['loss'].iloc[min_loss_index]

        # Mark minimum loss point
        ax.scatter(min_loss_iteration, min_loss, color=color, s=50, marker='o')

        # Annotate minimum loss value
        ax.annotate(f'{min_loss:.2f}', (min_loss_iteration, min_loss), textcoords="offset points", xytext=(5, 5),
                    ha='center', fontsize=11)

    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_ylabel('Loss', fontsize=12)
    ax.set_title(title, fontsize=18, fontname='Arial')
    ax.set_xlim(0, 2000)
    # ax.grid(True)  # Add grid lines
    ax.legend(loc='upper right', fontsize=10)  # Place legend outside the plot

    plt.tight_layout()
    plt.show()

# Plot for 2S logs
plot_logs(log_files_2S, 'Loss per Iteration for ESM1bts + ECFP (2S)')

# Plot for 3S logs
plot_logs(log_files_3S, 'Loss per Iteration for ESM1bts + ECFP (3S)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import re

def parse_log(file_path):
    with open(file_path, 'r') as file:
        log_lines = file.readlines()
    iteration_data = []
    unique_losses = set()
    for line in log_lines:
        match_iteration = re.search(r'Iteration (\d+)', line)
        match_loss = re.search(r'Best loss so far: (\d+\.\d+)', line)
        if match_iteration:
            current_iteration = int(match_iteration.group(1))
        elif match_loss:
            current_loss = float(match_loss.group(1))
            # Check for unique loss in the same iteration
            if current_loss not in unique_losses:
                iteration_data.append({'iteration': current_iteration, 'loss': current_loss})
                unique_losses.add(current_loss)
    return pd.DataFrame(iteration_data)

log_directory = '/Users/vahidatabaigi/SIP/data/Reports/hyperOp_report'
log_files = [f for f in os.listdir(log_directory) if f.endswith('.log')]

color_map = {'ESP(1D)': 'red', 'C1e': 'blue', 'C1f': 'green', 'I1e': 'black', 'I1f': 'magenta'}

fig, ax = plt.subplots(figsize=(10, 6))

for log_file in log_files:
    experiment_name = os.path.splitext(log_file)[0].split('Hyperparameter_optimization_ESM1bts_and_PreGNN_')[1]
    color = color_map.get(experiment_name, 'gray')
    log_data = parse_log(os.path.join(log_directory, log_file))
    
    # Smooth the data using moving average
    window_size = min(20, max(1, len(log_data['iteration']) // 10))  # Adjust window size as needed
    smooth_loss = np.convolve(log_data['loss'], np.ones(window_size)/window_size, mode='same')
    smooth_iteration = log_data['iteration']

    # Convert smooth_iteration and smooth_loss to Pandas Series
    smooth_iteration = pd.Series(smooth_iteration)
    smooth_loss = pd.Series(smooth_loss)

    # Extend the curve to iteration 2000
    last_iteration = smooth_iteration.iloc[-1]
    remaining_iterations = np.arange(last_iteration + 1, 2001)
    extension_loss = np.full_like(remaining_iterations, smooth_loss.iloc[-1])
    smooth_iteration = smooth_iteration.append(pd.Series(remaining_iterations))
    smooth_loss = smooth_loss.append(pd.Series(extension_loss))

    # Plot smoothed data
    ax.plot(smooth_iteration, smooth_loss, color=color, label=experiment_name, alpha=0.8)

    # Find minimum loss point from original data
    min_loss_index = log_data['loss'].idxmin()
    min_loss_iteration = log_data['iteration'].iloc[min_loss_index]
    min_loss = log_data['loss'].iloc[min_loss_index]
    
    # Mark minimum loss point
    ax.scatter(min_loss_iteration, min_loss, color=color, s=50, marker='o')
    
    # Annotate minimum loss value
    ax.annotate(f'{min_loss:.2f}', (min_loss_iteration, min_loss), textcoords="offset points", xytext=(5,5), ha='center', fontsize=11)

ax.set_xlabel('Iteration', fontsize=12)
ax.set_ylabel('Loss', fontsize=12)
ax.set_title('Loss per Iteration for ESM1bts + PreGNN', fontsize=18, fontname='Arial')
ax.set_xlim(0, 2000)
#ax.grid(True)  # Add grid lines
ax.legend(loc='upper right', fontsize=10)  # Place legend outside the plot

plt.tight_layout()
plt.show()
```
