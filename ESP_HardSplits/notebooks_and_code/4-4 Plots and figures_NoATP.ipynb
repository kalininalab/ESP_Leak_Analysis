```python
import matplotlib.pyplot as plt
from sklearn import metrics

def plot_roc_curve(fpr, tpr, roc_auc, label, linestyle='solid', color='blue'):
    plt.plot(fpr, tpr, color=color, label=f'{label}, AUC = {roc_auc:.2f}', linewidth=2.0, linestyle=linestyle)

fig, ax = plt.subplots(figsize=(8, 8))
plt.rcParams.update({'font.size': 20})

# Define your data and labels here
data = [

    (fpr_esm1b_ts_ecfp_C1e_D3408, tpr_esm1b_ts_ecfp_C1e_D3408, roc_auc_esm1b_ts_ecfp_C1e_D3408, 'C1e split, Del 3408, [ESM1bts + ECFP]', 'solid', 'blue'),
    (fpr_esm1b_ts_gnn_C1e_D3408, tpr_esm1b_ts_gnn_C1e_D3408, roc_auc_esm1b_ts_gnn_C1e_D3408, 'C1e split, Del 3408, [ESM1bts + PreGNN]', 'dashed', 'blue'),
    
    (fpr_esm1b_ts_ecfp_C1e_NOATP, tpr_esm1b_ts_ecfp_C1e_NOATP, roc_auc_esm1b_ts_ecfp_C1e_NOATP, 'C1e split, NO ATP, [ESM1bts + ECFP]', 'solid', 'black'),
    (fpr_esm1b_ts_gnn_C1e_NOATP, tpr_esm1b_ts_gnn_C1e_NOATP, roc_auc_esm1b_ts_gnn_C1e_NOATP, 'C1e split, NO ATP [ESM1bts + PreGNN]', 'dashed', 'black'),
    
    (fpr_esm1b_ts_ecfp_Rand_D3408, tpr_esm1b_ts_ecfp_Rand_D3408, roc_auc_esm1b_ts_ecfp_Rand_D3408, 'ESP split, Del 3408, [ESM1bts + ECFP]', 'solid', 'green'),
    (fpr_esm1b_ts_gnn_Rand_D3408, tpr_esm1b_ts_gnn_Rand_D3408, roc_auc_esm1b_ts_gnn_Rand_D3408, 'ESP split,Del 3408, [ESM1bts + PreGNN]', 'dashed', 'green'),
    
    (fpr_esm1b_ts_ecfp_Rand_NOATP, tpr_esm1b_ts_ecfp_Rand_NOATP, roc_auc_esm1b_ts_ecfp_Rand_NOATP, 'ESP split, No ATP, [ESM1bts + ECFP]', 'solid', 'red'),
    (fpr_esm1b_ts_gnn_Rand_NOATP, tpr_esm1b_ts_gnn_Rand_NOATP, roc_auc_esm1b_ts_gnn_Rand_NOATP, 'ESP split, No ATP, [ESM1bts + PreGNN]', 'dashed', 'red'),

]
# Plot each ROC curve
for fpr, tpr, roc_auc, label, linestyle, color in data:
    plot_roc_curve(fpr, tpr, roc_auc, label, linestyle, color)

# Set axis parameters
ax.locator_params(axis="y", nbins=5)
ax.locator_params(axis="x", nbins=5)

# Set legend
plt.legend(loc='lower right', bbox_to_anchor=(1.30, 0.01), fontsize=14)
plt.plot([0, 1], [0, 1], '--')
eps = 0.01
plt.xlim([0 - eps, 1 + eps])
plt.ylim([0 - eps, 1 + eps])
plt.ylabel('True Positive Rate',fontsize=16)
plt.xlabel('False Positive Rate',fontsize=16)
plt.show()
```
